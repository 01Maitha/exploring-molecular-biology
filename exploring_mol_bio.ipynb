<span class="emoji"><font size=5> üîé</font></span><font color=#F36F6F size=4><b> What is codon? </b></font>

<font color="#5AC2D3"><b>Codon</b></font>. It is the DNA language, that consists of 64 words of 4 letters only, that represent 4 individual molecules called **nucleotides**: thymine (T), adenine (A), cytosine (C), and guanine (G). However, thymine (T) in RNA is replaced by the nucleotide base uracil (U).  And so, a codon is a sequence of three nucleotides on a strand of DNA or RNA.

A codon includes **stop codons** that indicates the end of the protein instructions. The three stop codons in RNA are UAG, UAA, and UGA. It also includes a **start** codon which is AUG. <a href="https://www.khanacademy.org/science/biology/gene-expression-central-dogma/central-dogma-transcription/a/the-genetic-code-discovery-and-properties">(4)</a>

```python
#----Translation Function------#
translation = {}
for a in codon.split("\n"):
    k,v= a.split("\x09")
    if '/' in k:
        k= k.split("/")[-1].strip()
    ##ADDED to replace STOP with #
    k= k.replace("STOP", "#")
    v= v.replace(",", "").replace(";", "").lower().replace("u","t").split(" ")  
    for vv in v:
        if vv in translation:
            print("dup", vv) 
        translation[vv]= k
```

```python
#----Translation Function------#

def translate(x, protein = False):
    aa = []
    for i in range(0, len(x)-2, 3):
        aa.append(translation[x[i:i+3]])  
    aa = ''.join(aa)
    if protein:
        assert aa[0]== "M"
        assert aa[-1] == "#"
        aa = aa[:-1]
    return aa

#decode reading frame 0, 1, and 2 
translated_codon = translate(sars2[0:]) + translate(sars2[1:]) + translate(sars2[2:])
```

<span class="emoji"><font size=5> üîé</font></span><font color=#F36F6F size=4><b> What is Edit Distance? </b></font> 

**Edit distance**. A measure for the similarity between two strings. And it is the minimum number of change, insert or delete operations that transform one string to the other. **Types** There are  popular measures of edit distance, which are calculated using a different set of allowable edit operations. For instance: <font color="#5AC2D3"><b>Damerau‚ÄìLevenshtein distance</b></font>, Longest Common Subsequence (LCS) distance, Hamming distance, and Jaro distance.


<font color="#5AC2D3"><b>Damerau‚ÄìLevenshtein distance</b></font> operations are:
- Substituting a character,
- Deleting a character,
- Inserting a character, or
- Swapping two characters in the original strings
    
**In Bioinformatics**, the <font color="#5AC2D3"><b>Damerau‚ÄìLevenshtein distance</b></font> is an appropriate metric of the variation between two strands of DNA used to find the minimum number of substitution required to transform one sequence into another <a href="https://en.wikipedia.org/wiki/Damerau%E2%80%93Levenshtein_distance#DNA">(2)</a>. And it is an efficient algorithm used for rapid DNA sequence matching operation <a href="https://pdfs.semanticscholar.org/85fe/354eba01672f1c16413f16de3925c039d48e.pdf">(3)</a>.

The genome organization for a coronavirus is 5‚Ä≤-leader-UTR-replicase **(ORF1ab)**- **spike (S)** - **envelope (E)** - **membrane (M)** -**nucleocapsid (N)**-3‚Ä≤UTR-poly (A) tail.
